```python
# W6A1
# Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
# giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
# Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

lst = map(int, input().split())
out = []

for i in lst:
    if i not in out:
        out.append(i)
        
print(out)
```

```python
# W6A2
# Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
# tới vị trí i), giữ nguyên thứ tự

lst = map(int, input().split())
s = 0

for i in lst:
    s += i
    print(s, end = " ")
    
```

```python
# W6A3
# Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
# sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
# xếp.)

tup = tuple(map(int, input().split()))
k = int(input())
n = len(tup)

for i in range (k, n):
    print(tup[i], end = " ")
    
for i in range(k):
    print(tup[i], end = " ")
    
```

```python
# W6A4
# Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. 
# Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. 
# Nếu key lặp lại, thêm vào list của key đó

dct = dict()
lst = list(map(str,input().split()))
lst_key = []
lst_value = []
s = []

for i in lst:
    k = i.split(":")
    if (k[0] not in s):
        s.append(k[0])
    lst_key.append(k[0])
    lst_value.append(int(k[1]))   

for i in s:
    value = []
    for j in range (len(lst_key)):
        if (i == lst_key[j]):
            value.append(lst_value[j])
            
    dct.update({i : value})
    
print(dct)
```

```python
# W6A5
# Nhập một list số nguyên, In ra dictionary 
# {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

lst = list(map(int, input().split()))
pos = 0
neg = 0
zero = 0

for i in lst:
    if (i > 0):
        pos += 1
    elif (i < 0):
        neg += 1
    else:
        zero += 1
        
dct = dict()
dct.update({'positives' : pos})
dct.update({'negatives' : neg})
dct.update({'zero' : zero})

print(dct)
```

```python
# W6A7
# Viết chương trình nhập vào một tuple, in ra phần tử đầu, phần tử cuối, và
# tuple đảo ngược

tup = tuple(map(int, input().split()))
n = len(tup)

print(f"Phan tu dau: {tup[0]}")
print(f"Phan tu cuoi: {tup[n-1]}")
for i in range (n-1,-1, -1):
    print(tup[i], end = " ")
    
```

```python
# W6A8
# Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện
# của từng chuỗi và lưu vào dictionary

lst = list(map(str, input().split()))
lst_key = []
dct = dict()

for i in lst:
    if (i not in lst_key):
        lst_key.append(i)

for i in lst_key:
    s = 0
    for j in lst:
        if (i == j):
            s += 1
    
    dct.update({i : s})
    
print(dct)
```

```python
# W6A9
# Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
# dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
# được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
# theo thứ tự từ điển tăng dần của key

lst1 = list(map(str, input().split()))
lst2 = list(map(str, input().split()))
dct = dict()

lst_key = []
for i in lst1:
    k = i.split(':')
    lst_key.append(k[0])
    dct.update({k[0] : int(k[1])})
    
for i in lst2:
    k = i.split(':')
    if (k[0] not in lst_key):
        dct.update({k[0] : int(k[1])})
    else:
        dct[k[0]] += int(k[1])

sorted_key = sorted(dct.items())
sorted_dct = dict(sorted_key)
print(sorted_dct)
    
```

```python
# W6A10
# Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
# dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
# một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
# sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
# sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

lst = list(map(int, input().split()))
lst_out = []
k = int(input())

lst.sort()
for i in lst:
    if (i > (k-i)):
        break    
    
    if ((k - i) in lst) and ((i, k-i) not in lst_out):
        lst_out.append((i, k-i))

    if (2 * (i+1) == k):
        if (lst.count(i+1) >= 2):
            lst_out.append((i+1, i+1))
        break
    
print(lst_out)
    
```

```python
# W6A11
# [Tách số chẵn và lẻ từ tuple]
# Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
# tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
# thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt

tup = tuple(map(int, input().split()))

chan = []
le = []

for i in tup:
    if (i % 2 == 0):
        chan.append(i)
    else:
        le.append(i)
        
print(f"tuple chua so chan: {tuple(chan)}")
print(f"tuple chua so le: {tuple(le)}")
```

```python
# W6A12
# [Tìm phần tử xuất hiện nhiều nhất]
# Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
# list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
# nhất trong số đó.

lst = list(map(int, input().split()))
lst_cnt = []
lst.sort()

dct = dict()

for i in lst:
    if (str(i) not in dct):
        dct.update({str(i) : lst.count(i)})
        lst_cnt.append(lst.count(i))
        
k = max(lst_cnt)
 
for i in dct:
    if (dct[i] == k):
        print(i)
        break
```

```python
# W6A13
# [Đảo ngược key và value của dictionary]
# Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
# một dictionary mới bằng cách đảo ngược key và value của dictionary ban
# đầu. In ra dictionary mới.

lst = list(map(str, input().split()))

dct = dict()

for i in lst:
    k = i.split(":")
    dct.update({k[1] : k[0]})
    
print(dct)
```

```python
# W6A14 [Giao của hai danh sách]
# Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. 
# Hãy tìm cácphần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
# phần tử chung đó, không trùng lặp.

lst_1 = set(map(int, input().split()))
lst_2 = list(map(int, input().split()))

for i in lst_1:
    if (i in lst_2):
        print(i, end = " ")
        
```

```python
# W6A15
# [Lọc dictionary theo giá trị]
# Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
# nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
# key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
# mới.

lst = list(map(str, input().split()))
k = int(input())
dct = dict()

for i in lst:
    n = i.split(':')
    if (int(n[1]) > k):
        dct.update({n[0] : int(n[1])})
        
print(dct)
    
```
