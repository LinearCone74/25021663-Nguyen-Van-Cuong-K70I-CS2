```python
# W5A1
# Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two(
# a, b) -> int:

def Max_of_Two (a, b):
    if (a > b):
        return a
    return b

a, b = map(int, input().split())
print(Max_of_Two(a, b))
```

```python
# W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]

def swap(a, b):
    a, b = b, a
    return (a, b)

a, b = map(int, input().split())

print(swap(a, b))
```

```python
# W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

def snt (n):
    if (n < 2): return False
    
    if ((n == 2) or (n == 3)): return True
    
    for i in range (2, int(n**0.5) + 1):
        if (n % i == 0): 
            return False
        
    return True

n = int(input())

print(snt(n))
        
```

```python
# W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

def shh (n):
    s = 1
    for i in range (2,int(n**0.5) + 1):
        if (n % i == 0):
            s += i
            if (n / i != i):
                s += n // i
    
    if (s == n):
        return True
    
    return False

n = int(input())
print(shh(n))
```

```python
# W5A5
# Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
# phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

def ans (lst, k):
    for i in range (len(lst)):
        if (lst[i] == k):
            return i + 1
        
    return -1

lst = list(map(int, input().split()))
k = int(input())

print(ans(lst, k))
```

```python
# W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

def gt (n):
    if ((n == 0) and (n == 1)): return 1
    
    s = n
    for i in range (1,n):
        s *= i
    
    return s

n = int(input())
print(gt(n))
```

```python
# W5A7
# Viết hàm giả lập một máy tính bỏ túi.
# Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử
# \( operat \)
# Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
# Phép trừ \( - \)
# Phép cộng \( + \)
# Phép chia \( / \)
# Phép nhân \( * \)
# Kết quả làm tròn đến chữ số thập phân thứ hai.
# Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
# Đầu ra: Kết quả phép toán.

def casio (s):
    if ("+" in s):
        a, b = map(float, s.split("+"))
        return round(a+b, 3)
    
    elif ("-" in s):
        a, b = map(float, s.split("-"))
        return round(a-b, 3)
    
    elif ("*" in s):
        a, b = map(float, s.split("*"))
        return round(a*b, 3)
    
    else:
        a, b = map(float, s.split("/"))
        return round(a/b, 3)
    
s = input()
print(f"{s} = {casio(s)}")
```

```python
# W5A8
# Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits
# tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
# Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
# Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu
# lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (errordetecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính
# số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để
# ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi
# là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn
# được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã
# học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu
# hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào,
# hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
# Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

def bits (n):
    lst = []
    while (n > 0):
        if (n % 2 == 0):
            lst.insert(0, 0)
        else:
            lst.insert(0, 1)
        
        n = n // 2
        
    return lst

def hamming (a, b):
    lst_a = bits(a)
    lst_b = bits(b)
    
    s = abs(len(lst_a) - len(lst_b))
    
    for i in range (min(len(lst_a), len(lst_b))):
        if lst_a[i] != lst_b[i]:
            s += 1
    
    return s

a, b = map(int, input().split())

print(hamming(a, b))
        
```

```python
# W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n

def tong_cs (n):
    s = 0
    while (n > 0):
        s += n % 10
        n = n // 10
        
    return s
n = int(input())
print(tong_cs(n))
```

```python
# W5A10
# Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn
# lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ
# tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí
# tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng
# cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'.
# Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể
# giả sử hai từ đầu vào có cùng độ dài.

def axa (s):
    a = {" "}
    for i in s:
        a.add(i)
        
    return len(a) - 1

def dang_cau(s1, s2):
    if (axa(s1) == axa(s2)):
        return True
    
    return False

s1, s2 = map(str, input().split())

if (dang_cau(s1, s2)):
    print("True")
else:
    print("False")
     
```
