```python
# W7A1: 

def binary_search(arr, target):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] < target:
            left = mid + 1
        elif arr[mid] > target:
            right = mid - 1
        else:
            return mid
    return -1

n = int(input())
for _ in range(n):
    arr = list(map(int, input().strip().split()))
    target = int(input())
    print(binary_search(arr, target))

```

```python
# W7A2: 
def count_occurences(arr,x) -> int:
    cnt = 0
    for i in arr:
        if i == x:
            cnt += 1
    return cnt

n = int(input())
for _ in range(n):
    arr = list(map(int, input().strip().split()))
    x = int(input())
    print(count_occurences(arr,x))
```
