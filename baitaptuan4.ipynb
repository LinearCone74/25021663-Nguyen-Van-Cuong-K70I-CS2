```python
# W4A2
# [PrimeNumber]
# VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
# Kiểm tra số này có phải là số nguyên tố hay không?

def snt (n):
    if (n < 2): return True
    
    if ((n == 2) or (n == 3)): return True
    
    for i in range (2, int(n ** 0.5) + 1):
        if (n % i == 0):
            return False
    
    return True

n = int(input("Nhap n: "))
while (n < 0):
    n = int(input("Vui long nhap lai: "))
    
if (snt(n)):
    print(f"{n} la mot so nguyen to")
else:
    print(f"{n} khong la so nguyen to")
    
```

```python
# W4A3
# [Factorial]
# Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n)
# với 0 < n < 100.

def gt (n):
    if (n == 0): return 1
    s = 1
    for i in range(1, n+1):
        s *= i
        
    return s

n = int(input("Nhap n: "))
print(f"{n}! = {gt(n)}")
```

```python
# W4A6
# "[SumPrimesInRange]
# Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
# [a, b]. Mỗi dòng là một bộ test (a b)."

n = 10 ** 6
lst = n * [0]
lst[0] = 1
lst[1] = 1
i = 2
lst_test = []

while (i < n):
    if (lst[i] == 0):
        j = 2 * i
        while (j < n):
            lst[j] += 1
            j += i
    
    i += 1
    
n = int(input())
s = 0    

for i in range (n):
    a, b = map(int, input().split())

    for i in range (a, b+1):
        if (lst[i] == 0):
            s += i
            
    lst_test.append(s)
    s = 0
        
print(lst_test)
```

```python
# W4A8
# ReverseAddPalindrome]
# Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
# đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

def rev (n):
    s = 0
    while (n > 0):
        s = s * 10 + n%10
        n = int(n/10)
        
    return s

n = int(input())

while (n != rev(n)):
    n = n + rev(n)
    
print(n)
```

```python
# W4A9
# [PerfectSquaresDistinctDigits]
# Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
# đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
# cách). Nếu không có, in ra (no number)

def cs (n):
    lst = 10 * [0]
    while (n > 0):
        lst[n % 10] += 1
        n = n // 10
    
    for i in range(10):
        if (lst[i] > 1): return False
    
    return True
n = int(input())

for i in range(int(n**0.5)+1):
    if (cs(i*i)): 
        print(i*i, end = " ")
        
```

```python
# W4A10
# "[LongestCollatzUpToN]
# Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
# và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
# nhỏ nhất.

def collatz(n):
    cnt = 0
    while (n != 1):
        cnt += 1
        if (n % 2 == 0):
            n = n // 2
        else:
            n = 3*n + 1
    
    return cnt

lst = []
n = int(input())

for i in range (1, n):
    lst.append(collatz(i))
    
print(max(lst))
```

```python
# W4A13
# [ClosePairNumber]
# Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
# Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
# Ví dụ 220 và 284 là một cặp số thân thiết
# 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. 
# Tổng của 11 ước này cho chúng ta 284
# 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
# vừa đúng bằng 220
# Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
# Nếu có, in ra màn hình "true", ngược lại, in ra "false"

def tong_uoc(n):
    s = 1
    for i in range (2,int(n ** 0.5) + 1):
        if (n % i == 0):
            s += i
            if (n / i != i):
                s += int(n / i)
    
    return s

a, b = map(int, input().split())

if ((a == tong_uoc(b)) and (b == tong_uoc(a))):
    print("True")
else:
    print("False")

```

```python
# W4A14
# [GCD]
# Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. 
# Viết chương trình tìm ước chung lớn nhất của hai số m và n.

m, n = map(int, input().split())

if (m > n):
    m, n = n, m
    
for i in range(m, 0, -1):
    if ((n % i == 0) and (m % i == 0)):
        print(i)
        break
```

```python
# W4A15
# [ChickenAndDog]
# "Vừa gà vừa chó,
# Bó lại cho tròn,
# Ba mươi sáu con,
# Một trăm chân chẵn"
# Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
# con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
# số lượng chó. Ngược lại, in ra "invalid".

m, n = map(int, input().split())

if (n % 2 != 0):
    print("invalid")
elif (n - 2*m < 0):
    print("invalid")
else:
    dog = (n - 2*m) // 2
    chicken = m - dog
    print(f"so ga la {chicken}")
    print(f"so cho la {dog}")
    
```
